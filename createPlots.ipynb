```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools

# load data -- file deliniated by space
file_path = 'data/flow-cytometry-40k.txt'
data = pd.read_csv(file_path, delim_whitespace=True, header=None, usecols=range(1, 6))
# columns according to the dimensions
dimensions = [1, 2, 3, 4, 5]

# get all combinations of dimensions
combinations = list(itertools.combinations(dimensions, 2))

def get_limits(series, lower=0.01, upper=0.99):
    lower_limit = series.quantile(lower)
    upper_limit = series.quantile(upper)
    return lower_limit, upper_limit

# plot all combinations
for combo in combinations:
    print(combo)
    x_upper = 500
    y_upper = 500
    _, x_upper = get_limits(data[combo[0]])
    _, y_upper = get_limits(data[combo[1]])
    plt.scatter(data[combo[0]], data[combo[1]], s=0.3)
    # plt.xlabel(combo[0])
    # plt.ylabel(combo[1])
    # plt.title(f'{combo[0]} vs {combo[1]}')
    plt.axis('off')
    plt.xlim(0, x_upper)
    plt.ylim(0, y_upper)
    plt.savefig(f'gen_images/{combo[0]}_vs_{combo[1]}.png', bbox_inches='tight', pad_inches=1, transperant=True)
    plt.show()
```
