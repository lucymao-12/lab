```python
PATH_TO_IMAGES = "feb15/fcs_png/polys_sol_fill"
```

```python
import os
if os.path.exists(PATH_TO_IMAGES):
    print("Path exists")
else:
    print("Path does not exist")
```

```python
## get all folder name in directory
## folders are names of the scatter plot from flow cytometry data
folders = [f for f in os.listdir(PATH_TO_IMAGES) if os.path.isdir(os.path.join(PATH_TO_IMAGES, f))]
print(folders)
```

```python
## map: folder name : list of images in the folder
image_dict = {}
for folder in folders:
    image_dict[folder] = [f for f in os.listdir(os.path.join(PATH_TO_IMAGES, folder)) 
                          if os.path.isfile(os.path.join(PATH_TO_IMAGES, folder, f)) 
                          and f.endswith(".png")]
print(image_dict)
```

```python
from PIL import Image
```

```python
def adjust_transparency(image, alpha_value):
    """ Adjust the alpha channel of an image to the specified alpha value (0-255) """
    r, g, b, alpha = image.split()  # Split the image into its component channels
    new_alpha = alpha.point(lambda p: alpha_value)  # Set alpha channel to alpha_value
    return Image.merge('RGBA', (r, g, b, new_alpha))
```

```python
# create image overlays of all images for each folder
for folder in image_dict:
    print(folder)
    transparency_coeff = len(image_dict[folder])
    alpha = (255 // transparency_coeff) + 1
    images = [Image.open(os.path.join(PATH_TO_IMAGES, folder, image)) for image in image_dict[folder]]
    adjusted_images = [adjust_transparency(image, alpha) for image in images]
    base = adjusted_images[0]
    for overlay in adjusted_images[1:]:
        base = Image.alpha_composite(base, overlay)
    base.save(os.path.join("blended/" + folder + ".png"), "PNG")
    
    
```

```python
PATH_TO_PLOTS = "fcs_png/pics"
# get names for all images in this folder
images_names = [f for f in os.listdir(PATH_TO_PLOTS) if os.path.isfile(os.path.join(PATH_TO_PLOTS, f)) and f.endswith(".png")]
image_no_ext = [f.split(".")[0] for f in images_names]

PATH_TO_OVERLAYS = "blended"
# get names for all images in this folder
overlay_names = [f for f in os.listdir(PATH_TO_OVERLAYS) if os.path.isfile(os.path.join(PATH_TO_OVERLAYS, f)) and f.endswith(".png")]
overlay_no_ext = [f.split(".")[0] for f in overlay_names]
# compare if any image names are the same
count = 0
total = 0
for image in image_no_ext:
    if image in overlay_no_ext:
        print(image)
        count += 1
        total += 1
    else:
        print("No match found for " + image)
        total += 1
print("Total images: " + str(total))
print("Images with overlay: " + str(count))
```

```python
##get the darkest region of the image
def get_darkest_region(image, threshold):
    """ Get the darkest region of the image """
    image = image.convert("RGBA")
    data = image.getdata()
    dark = []
    for item in data:
        brightness = 0.299 * item[0] + 0.587 * item[1] + 0.114 * item[2]
        if brightness < threshold:
            dark.append((0, 0, 0, 255))
        else:
            dark.append((255, 255, 255, 0))
    returned = Image.new("RGBA", image.size)
    returned.putdata(dark)
    return returned
```

```python
for image in overlay_no_ext:
    img = Image.open(os.path.join("blended/", image + ".png"))
    darkest = get_darkest_region(img, 20)
    darkest.save(os.path.join("darkest/" + image + ".png"), "PNG")
    
```
